```python
import random

def print_board(board):
    for row in board:
        print(" ".join(row))
    print()

def calculate_attacks(board):
    n = len(board)
    attacks = 0
    for i in range(n):
        for j in range(n):
            if board[i][j] == 'Q':
                # Check horizontal and vertical attacks
                for k in range(n):
                    if board[i][k] == 'Q' and k != j:
                        attacks += 1
                    if board[k][j] == 'Q' and k != i:
                        attacks += 1
                # Check diagonal attacks
                for k in range(1, n):
                    if i+k < n and j+k < n and board[i+k][j+k] == 'Q':
                        attacks += 1
                    if i+k < n and j-k >= 0 and board[i+k][j-k] == 'Q':
                        attacks += 1
                    if i-k >= 0 and j+k < n and board[i-k][j+k] == 'Q':
                        attacks += 1
                    if i-k >= 0 and j-k >= 0 and board[i-k][j-k] == 'Q':
                        attacks += 1
    return attacks // 2  # Each queen is counted twice

def hill_climbing(n):
    board = [['.' for _ in range(n)] for _ in range(n)]
    for i in range(n):
        j = random.randint(0, n-1)
        board[i][j] = 'Q'
    
    print("Initial Board:")
    print_board(board)
    
    attacks = calculate_attacks(board)
    print("Initial Attacks:", attacks)
    
    while True:
        min_attacks = attacks
        for i in range(n):
            for j in range(n):
                if board[i][j] == 'Q':
                    for k in range(n):
                        if k != j:
                            board[i][j], board[i][k] = '.', 'Q'
                            new_attacks = calculate_attacks(board)
                            if new_attacks < min_attacks:
                                min_attacks = new_attacks
                                best_move = (i, k)
                            board[i][j], board[i][k] = 'Q', '.'
        if min_attacks == attacks:
            break
        else:
            i, j = best_move
            board[i][j] = 'Q'
            attacks = min_attacks
            print("Board after move:")
            print_board(board)
            print("Attacks:", attacks)
    
    return board

n = 8  # Change this to any positive integer
solution = hill_climbing(n)
print("Final Solution:")
print_board(solution)
```
